```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].replace({'B': 0, 'M': 1}).astype(int)
df.head()
```

```python
df['diagnosis'].value_counts()
```

```python
target = 'diagnosis'
features = list(df.columns)
features.remove('diagnosis')
features
```

```python
X = df[features]
y = df[[target]]
```

Без масштабирования:

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values.reshape(-1), train_size=0.8, shuffle=True)
clf = LogisticRegression()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

С масштабированием:

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values.reshape(-1), train_size=0.8, shuffle=True)
clf = LogisticRegression()
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
y_pred = clf.predict_proba(X_test)
```

```python
df_results = pd.DataFrame({
    'pred': clf.predict(X_test).reshape(-1),
    'pred_proba': clf.predict_proba(X_test)[:, 1],
    'true': y_test.reshape(-1),
})
```
